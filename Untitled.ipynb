```python
import pandas as pd
import joblib
```

```python
df=pd.read_csv('dataset.csv', encoding='latin1')
df.head()
```

```python
import re
def get_unique_countries(location_list):
    countries = set()
    for location in location_list:
        match = re.search(r'(.*),(.*)', location)
        if match:
            countries.add(match.group(2).strip())
    return countries

countries=get_unique_countries(df['location'])
df['location'] = df['location'].str.extract(r'(.*),(.*)', expand=True)[1].str.strip()

# Replacing the countries that are not in the countries set with 'Other'
df['location'].replace(to_replace=list(set(df['location'])-countries), value="Other", inplace=True)
df['location'].head()
```

```python
final_dataset=df[['Rating', 'Total_Reviews', 'location', 'Skill_info_1',
       'Skill_info_2', 'Skill_info_3']]
final_dataset.head()
```

```python
from sklearn import preprocessing
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
final_dataset['Skill_info_1']=label_encoder.fit_transform(final_dataset['Skill_info_1'])
final_dataset['Skill_info_2']=label_encoder.fit_transform(final_dataset['Skill_info_2'])
final_dataset['Skill_info_3']=label_encoder.fit_transform(final_dataset['Skill_info_3'])
final_dataset['location']=label_encoder.fit_transform(final_dataset['location'])
```

```python
X= final_dataset.iloc[:,0:]
X.head()
```

```python
y=df.iloc[:,7]
y.head()
```

```python
#random forest regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=20)
regressor = RandomForestRegressor(n_estimators = 100, random_state = 30)
regressor.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
y_pred=regressor.predict(X_test)
y_pred
# calculate MSE and MAE
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mse",mse)
print("mae",mae)
print("r2",r2)
```

```python
import pickle
# open a file, where you ant to store the data
file = open('Predict.pkl', 'wb')
# dump information to that file
joblib.dump(regressor, file,compress=True)
```
